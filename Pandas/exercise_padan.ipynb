### ข้อมูลในตารางนี้ เป็นผู้โดยสารของเรือ Titanic ซึ่งมาจาก ชุดข้อมูล Titanic 
เป็นข้อมูลของผู้โดยสารทั้งหมด 891 คน ที่อยู่บนเรือ Titanic รวมทั้งข้อมูลส่วนบุคคลและสถานะการรอดชีวิต เพื่อใช้ในการวิเคราะห์หรือพยากรณ์ว่าใครรอดชีวิตจากอุบัติเหตุเรืออัปปาง

| คอลัมน์         | ความหมาย                                                            |
| --------------- | ------------------------------------------------------------------- |
| **PassengerId** | หมายเลขประจำตัวผู้โดยสาร                                            |
| **Survived**    | สถานะการรอดชีวิต (0 = ไม่รอด, 1 = รอดชีวิต)                         |
| **Pclass**      | ชั้นโดยสาร (1 = ชั้นหนึ่ง, 2 = ชั้นสอง, 3 = ชั้นสาม)                |
| **Name**        | ชื่อผู้โดยสาร                                                       |
| **Sex**         | เพศของผู้โดยสาร (male = ชาย, female = หญิง)                         |
| **Age**         | อายุของผู้โดยสาร                                                    |
| **SibSp**       | จำนวนพี่น้อง/คู่สมรสที่โดยสารมาด้วยกัน                              |
| **Parch**       | จำนวนพ่อแม่/ลูกที่โดยสารมาด้วยกัน                                   |
| **Ticket**      | หมายเลขตั๋ว                                                         |
| **Fare**        | ราคาตั๋ว (ค่าโดยสาร)                                                |
| **Cabin**       | หมายเลขห้องโดยสาร (บางคนไม่มีข้อมูล)                                |
| **Embarked**    | ท่าเรือที่ขึ้นเรือ (S = Southampton, C = Cherbourg, Q = Queenstown) |


```python
df_from_csv.columns

# เข้าถึงข้อมูล คอลัมน์ ‘Name’ ทั้งหมด
df_from_csv["Name"]

# เข้าถึงข้อมูล คอลัมน์ ‘Name’ และ ‘Age’ ของผู้โดยสารทุกคน
df_from_csv[["Name","Age"]]
```

```python
# เข้าถึงข้อมูลของผู้โดยสารที่ มี PassengerId = 50 (ทุกคอลัมน์)
Titanic = df_from_csv
Titanic_PassengerId_50 = df_from_csv[df_from_csv["PassengerId"]==50]
Titanic_PassengerId_50
```

```python
# เข้าถึงข้อมูลของผู้โดยสาร 5 คนแรก เฉพาะคอลัมน์ ‘Name’, ‘Age’, ‘Sex’
Titanic = df_from_csv
Titanic.loc[0:4,["Name","Age","Sex"]]
```

```python
# เข้าถึง ชื่อ (Name) และ ค่าโดยสาร (Fare) ของผู้โดยสารแถวที่ 100–110
Titanic.loc[99:109, ["Name", "Fare"]]
```

```python
#  เข้าถึงข้อมูล ผู้โดยสารที่รอดชีวิต (Survived = 1) เฉพาะคอลัมน์ ‘Name’ และ ‘Pclass’
Titanic[Titanic['Survived'] == 1][['Name', 'Pclass']]
```

```python
# เข้าถึงข้อมูล ผู้โดยสารที่อยู่ชั้น Pclass = 1 เฉพาะคอลัมน์ ‘Name’, ‘Sex’, ‘Age’
Titanic[Titanic["Pclass"]== 1] [["Name","Sex","Age"]]
```

```python
# เข้าถึงข้อมูลแบบกำหนด index + column พร้อมกัน (เช่น แถวที่ 15–20 เฉพาะคอลัมน์ ‘Name’, ‘Age’, ‘Survived’)
```

## 3. การกรองข้อมูล (Filtering)
แบบฝึกหัด

```python
# เลือกเฉพาะข้อมูลผู้โดยสารที่ รอดชีวิต (Survived = 1)
Titanic[Titanic["Survived"] == 1]
```

```python
# เลือกเฉพาะข้อมูลผู้โดยสารที่ อายุต่ำกว่า 18 ปี
Titanic_Age_more_than_18 = Titanic["Age"]<18
Titanic_Age_more_than_18


```

```python
# เลือกเฉพาะข้อมูลผู้โดยสารที่ เป็นผู้หญิง (Sex = ‘female’) และอยู่ชั้น Pclass = 1
Titanic[(Titanic["Sex"]=="female") & (Titanic["Pclass"] == 1)]

# f ต้องเป็นตัวพิมพ์เล็ก
```

```python
# เลือกเฉพาะผู้โดยสารที่ Fare มากกว่า 100
Titanic_fare_more_than_100 = Titanic["Fare"]>100
Titanic_fare_more_than_100

# แล้วทำไมอันนี้ F เป็นตัวพิมพ์ใหญ่ ?????????????????++++++
```

```python
# เลือกเฉพาะผู้โดยสารที่ อายุไม่เกิน 30 ปี และรอดชีวิต
Titanic[(Titanic["Age"] <= 30) & (Titanic["Survived"] == 1)]

# ยากมากกกกกกกกกกกกกกกกกก
```
