```python
import pandas as pd
import numpy as np 
from sklearn.model_selection import train_test_split  # แบ่งข้อมูลเป็นชุดฝึกสอนและทดสอบ
from sklearn.neighbors import KNeighborsClassifier # นำเข้าโมเดล KNN
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report # เมตริกสับสน , 
```

```python
# การใช้งาน Machine Learning
data = pd.read_csv('./dataset/Iris.csv')  # แสดง 5 แถวแรก
data.head()
```

```python
data.info()
```

```python
# กำหนด x และ y (ทาเกตหรือสปีชี้)
X = data[[ "SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]].values
y = data['Species']
```

```python
y
```

```python
X.shape
```

```python
# แบ่งชุดข้อมูลเป็นชุดฝึกสอน (80%) และชุดทดสอบ (20%)
# test_size= 0.2 หมายถึง 20% ของข้อมูลทั้งหมดจะถูกใช้เป็นชุดทดสอบ
# random_state= 42 ใช้เพื่อการแบ่งข้อมูลสามารถทำซ้ำได้
# X,y แบ่งข้อมูลคุณลักษณะและป้ายกำกับ

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size= 0.2, random_state= 1 )
# tam = 2
```

```python
X_train
```

```python
# เรียกใช้งาน โมเดล KNN 
knn_model = KNeighborsClassifier( n_neighbors = 5 )  # กำหนด K=3
# tam = 5
```

```python
# สอน โมเดลด้วยชุดฝึกสอน // train the model with training set
knn_model.fit(X_train, y_train)
```

```python
# test the model with testning set
y_pred = knn_model.predict(X_test)
```

```python
y_pred
```

```python
# ประเมินผลของโมเดลด้วยความแม่นยำ (Accuracy)
accuracy = accuracy_score(y_test, y_pred)
print(accuracy*100)
```

```python
Confusion_matx = confusion_matrix(y_test, y_pred)
print(Confusion_matx)
```

```python
clf_report = classification_report(y_test, y_pred)
print(clf_report)
```

```python
# Sumary
import pandas as pd
import numpy as np 
from sklearn.model_selection import train_test_split  # แบ่งข้อมูลเป็นชุดฝึกสอนและทดสอบ
from sklearn.neighbors import KNeighborsClassifier # นำเข้าโมเดล KNN
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report # เมตริกสับสน , 

# การใช้งาน Machine Learning
data = pd.read_csv('./dataset/Iris.csv')  # แสดง 5 แถวแรก
data.head()

data.info()

# กำหนด x และ y (ทาเกตหรือสปีชี้)
X = data[[ "SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]].values
y = data['Species']

# แบ่งชุดข้อมูลเป็นชุดฝึกสอน (80%) และชุดทดสอบ (20%)
# test_size= 0.2 หมายถึง 20% ของข้อมูลทั้งหมดจะถูกใช้เป็นชุดทดสอบ
# random_state= 42 ใช้เพื่อการแบ่งข้อมูลสามารถทำซ้ำได้
# X,y แบ่งข้อมูลคุณลักษณะและป้ายกำกับ

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size= 0.2, random_state= 1 )
# tam = 2

X_train

# เรียกใช้งาน โมเดล KNN 
knn_model = KNeighborsClassifier( n_neighbors = 5 )  # กำหนด K=3
# tam = 5

# สอน โมเดลด้วยชุดฝึกสอน // train the model with training set
knn_model.fit(X_train, y_train)

# test the model with testning set
y_pred = knn_model.predict(X_test)

y_pred

# ประเมินผลของโมเดลด้วยความแม่นยำ (Accuracy)
accuracy = accuracy_score(y_test, y_pred)
print(accuracy*100)

Confusion_matx = confusion_matrix(y_test, y_pred)
print(Confusion_matx)

clf_report = classification_report(y_test, y_pred)
print(clf_report)
```
