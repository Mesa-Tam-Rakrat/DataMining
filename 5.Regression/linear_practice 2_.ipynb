## 1. สร้างสมการถดถอยเชิงเส้น ของยอดขาย `Sales` จาก การโฆษณาผ่านทางหนังสือพิมพ์ `Newspaper`

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error , mean_squared_error , root_mean_squared_error
```

```python
# datasets 
tv_ad = pd.read_csv('./datatasest/advertising.csv')
tv_ad.head() # ตรวจสอบเบื้องต้น
```

```python
# Sale คู่กับ หนังสือพิมพ์ Newspaper
df = tv_ad[ ['Newspaper', 'Sales']]
df.head()
```

```python
# data preparation 
# 2 .MissingValue 
df.isnull().sum() 
```

```python
# Define x and y 
x = df ['Newspaper']    # type of x is series
y = df['Sales'] # type of y is series 

x = df ['Newspaper'].values.reshape(-1,1)    # values เอาแต่ค่าที่เป็นตัวเลข
y = df['Sales'].values.reshape(-1,1) 
```

```python
# data visualization with matplotlib (กราฟ)
plt.scatter(x, y, color = 'Orange', marker = 'o')
plt.xlabel('Advertise by Newspaper')
plt.ylabel('Sales Value')
plt.grid()
plt.show()
```

```python
# data visualization with seaborn 
sns.scatterplot( x = df['Newspaper'], y = df['Sales'] , color = 'orange' , data = df)  # scatterplot 
sns.lmplot( x= 'Newspaper', y = 'Sales' , data = df)  # scatterplot with regression line  เพิ่มเส้นมาด้วย
plt.show()
```

```python
# model training 
from sklearn.linear_model import LinearRegression
# Evaluation metrics 
from sklearn.metrics import mean_squared_error, mean_absolute_error #, r2_score
```

```python
# train model 

model = LinearRegression()
model.fit(x, y)
```

```python
# y = a + bx 
a = model.intercept_
b = model.coef_
print(a, b)
```

```python
# r2-score 
r_square = model.score(x, y) 
y_prediction = model.predict(x) # หาค่า y ทำนาย เพื่อนำไปหาค่า error 
mse = mean_squared_error(y, y_prediction)
mae = mean_absolute_error(y, y_prediction)
rmse = np.sqrt(mse) 
# rmse = np.sqrt(mse)
rmse_2 = root_mean_squared_error(y,y_prediction)
print(f"MSE: {mse : .4f}")
print(f"MAE: {mae : .4f}")
print(f"RMSE 1: {rmse : .4f}")
print(f"RMSE_2: {rmse_2 : .4f}")
```

```python
sns.set_style('darkgrid')   # white", "dark", "whitegrid", "darkgrid", "ticks"
sns.lmplot(x='Newspaper', y='Sales', data=df)
plt.show()
```

```python
# x = 300 ,  500, 1000 
# x= [ 300, 500, 1000]  ==> x = np.array(x).reshape(-1,1) 
x_input = [[300], [500], [ 1000]]  # list of list 
y_pred= model.predict(x_input)
y_pred
```
